```python
from langchain.document_loaders import DirectoryLoader

directory = '/content/sample_data'

def load_docs(directory):
  loader = DirectoryLoader(directory, show_progress=True, silent_errors=True)
  documents = loader.load()
  return documents
```

```python
documents = load_docs(directory)
```

```python
documents
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(documents,chunk_size=1000, chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs
```

```python
docs = split_docs(documents)
```

```python
import openai

from langchain.embeddings.openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model_name="ada")

query_result = embeddings.embed_query("Hello World")
```

```python
len(query_result)
```

```python
!pip3 install pinecone-client
import pinecone
from langchain.vectorstores import Pinecone

pinecone.init(
    api_key = "56789",
    environment="gcp-starter"
)

index_name = "microsoftesg"

index = Pinecone.from_documents(docs, embeddings, index_name=index_name)
```

```python
def get_similar_docs(query, k=2, score=False):
  if score:
    similar_docs = index.similarity_search_by_vector_with_score(query, k=k)
  else:
    similar_docs = index.similarity_search(query, k=k)
  return similar_docs

query = 'How much people will microsoft increase water access to by 2030'
similar_docs = get_similar_docs(query, k=2)
similar_docs
```
